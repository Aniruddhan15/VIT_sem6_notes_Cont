You are asked to design a machine learning model to decide which crop can be
grown in a particular farm based the farm characteristics such as rainfall, Nitrogen
and Phosphorous values of soil etc.
Apply decision tree algorithm for crop suggestion.
1. Compare the accuracy with both entropy and gini index for attribute
selection
2. Verify whether both metrics result in same decision tree by drawing tree
upto a depth level 2.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

data = pd.read_csv("Crop.csv")
target_variable = "label"
feature_variables = ["N", "P", "K", "temperature","humidity","rainfall"]
X_train, X_test, y_train, y_test = train_test_split(data[feature_variables], data[target_variable], test_size=0.2, random_state=42)
# Build decision trees
model_entropy = DecisionTreeClassifier(criterion="entropy", max_depth=2)
model_gini = DecisionTreeClassifier(criterion="gini", max_depth=2)
# Train models
model_entropy.fit(X_train, y_train)
model_gini.fit(X_train, y_train)
# Evaluate accuracy
accuracy_entropy = accuracy_score(y_test, model_entropy.predict(X_test))
accuracy_gini = accuracy_score(y_test, model_gini.predict(X_test))
print("Accuracy with Entropy:", accuracy_entropy)
print("Accuracy with Gini:", accuracy_gini)
# Plot and compare trees
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
# Entropy tree
plot_tree(model_entropy, ax=axes[0], feature_names=feature_variables, fontsize=10)
axes[0].set_title("Decision Tree with Entropy")
# Gini index tree
plot_tree(model_gini, ax=axes[1], feature_names=feature_variables, fontsize=10)
axes[1].set_title("Decision Tree with Gini Index")

plt.tight_layout()
plt.show()
```
